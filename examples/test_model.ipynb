### This is an example of how to test your trained model on a website screenshot

1. Save a fullscreen screenshot in Chrome by going to Tools -> More Tools -> Developer Tools and in mobile view click on the right menu and select "Capture full size screenshot"

Run test_net.py with the following arguments 
* model-name -> the name of the trained model
* model-file -> your trained model state_dict
* image-file -> the fullsize screenshot you took
* trained-with-residuals -> specify if the network was trained with residuals
* channel -> select which channel you want to view {'buttons': 0, 'forms': 1,                                      'images': 2, 'section': 3, 'textblock': 4} 


```python
!python3 test_net.py --model-name frednetv4 --model-file frednetv4.pth --image-file website_screenshot.png --trained-with-residuals y --channel 4
```
